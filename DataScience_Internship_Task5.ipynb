```python
# Problem 1
def analyze_post_engagement(likes_list):
    """
    Calculate total likes and classify post status.
    """
    total_likes = sum(likes_list)
    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")

# Example
analyze_post_engagement([200, 300, 550])
```

```python
#Problem 2
def check_medicine_stock(stock):
    """
    Check medicine stock levels.
    """
    status = "Low Stock Alert" if stock < 10 else "Stock Sufficient"
    print(f"Medicine Stock: {stock}")
    print(f"Status: {status}")

# Example
check_medicine_stock(6)
```

```python
#Problem 3
def rainfall_checker(rainfall_data, required_level=70):
    """
    Calculate average rainfall and check adequacy.
    """
    avg_rainfall = sum(rainfall_data) / len(rainfall_data)
    status = "Adequate Rainfall" if avg_rainfall >= required_level else "Inadequate Rainfall"
    print(f"Average Rainfall: {avg_rainfall}")
    print(f"Rainfall Status: {status}")

# Example
rainfall_checker([60, 80, 76])
```

```python
#Problem 4
def detect_duplicates(usernames):

    duplicates_found = len(usernames) != len(set(usernames))
    print(f"Duplicate Accounts Found: {'Yes' if duplicates_found else 'No'}")

# Example
detect_duplicates(["alice", "bob", "alice"])
```

```python
#Problem 5
def appointment_eligibility(age):
    """
    Check if patient is eligible for appointment.
    """
    status = "Eligible" if age >= 18 else "Not Eligible"
    print(f"Patient Age: {age}")
    print(f"Eligibility Status: {status}")

# Example
appointment_eligibility(21)
```

```python
#Problem 6
def premium_crop_filter(prices):
    """
    Filter crops above premium threshold.
    """
    premium = [p for p in prices if p > 2000]
    print(f"Premium Crops: {premium}")

# Example
premium_crop_filter([1500, 2500, 3200])
```

```python
#Problem 7
def app_health_checker(errors):
    """
    Evaluate application health based on error count.
    """
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    print(f"Error Count: {errors}")
    print(f"System Status: {status}")

# Example
app_health_checker(7)
```

```python
#Problem 8
def transaction_checker(amount, limit=50000):
    """
    Validate daily transaction amount.
    """
    status = "Approved" if amount <= limit else "Rejected"
    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")

# Example
transaction_checker(60000)
```

```python
#Problem 9
def attendance_checker(attendance_list, total_classes):
    """
    Calculate attendance percentage and check eligibility.
    """
    attended = sum(attendance_list)
    percentage = (attended / total_classes) * 100
    status = "Eligible" if percentage >= 75 else "Not Eligible"
    print(f"Attendance Percentage: {percentage:.1f}")
    print(f"Exam Eligibility: {status}")

# Example
attendance_checker([1,1,0,1,1,1,0,1], 10)
```

```python
#Problem 10
def electricity_bill(units):
    """
    Calculate electricity bill and classify usage.
    """
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print(f"Total Bill: {bill}")
    print(f"Usage Status: {usage}")

# Example
electricity_bill(250)
```
